```python
import os
import shutil
```

```python
#variaveis

caminho_inicial = "./arquivo_eleicao/"

# Questão 2
list_uf=['AC', 'AL', 'AM', 'AP', 'BA', 'CE', 'ES', 
         'GO', 'MA', 'MG', 'MS', 'MT', 'PA', 'PB', 
         'PE', 'PI', 'PR', 'RJ', 'RN', 'RO', 'RR', 
         'RS', 'SC', 'SE', 'SP', 'TO']


#Questão 3

dict_diretorio = {
    "bem_candidato": {
        "caminho": "./arquivo_eleicao/bem_candidato"
    },
    "consulta_candidato": {
        "caminho": "./arquivo_eleicao/consulta_cand"
    },
    "consulta_coligacao": {
        "caminho": "./arquivo_eleicao/consulta_coligacao"
    },
    "consulta_vagas": {
        "caminho": "./arquivo_eleicao/consulta_vagas"
    },
    "motivo_cassacao": {
        "caminho": "./arquivo_eleicao/motivo_cassacao"
    }
}
```

```python
#Questão 4


for chave, valor in dict_diretorio.items():
    os.makedirs(valor["caminho"], exist_ok=True)


for root, dirs, files in os.walk(caminho_inicial):
    for file in files:
        partes = file.split('_') 
        if len(partes) > 2:
            analise = partes[-2]  
            nome_arquivo = file.split(f'_{analise}')[0]
            
       
            for dir in dirs:
                if nome_arquivo == dir:
                    caminho_arquivo = os.path.join(root, file)
                    destino = os.path.join(caminho_inicial, nome_arquivo)
                    shutil.move(caminho_arquivo, destino)
                    print(f"Arquivo {file} movido para {destino}")
        else:
            print(f"O arquivo {file} não possui underscores suficientes para processamento.")
```

```python
#Questão 5

for chave, valor in dict_diretorio.items():
    for uf in list_uf:
        os.makedirs(f"{valor["caminho"]}/{uf}", exist_ok=True)
        for root, dirs, files in os.walk(valor["caminho"]):
            for file in files:  
                nome_arquivo = file.split('_')[-1].replace('.csv',"")
                for dir in dirs:
                    if nome_arquivo == dir:
                        caminho_arquivo = os.path.join(root, file)  
                        destino = os.path.join(valor["caminho"], nome_arquivo) 
                        shutil.move(caminho_arquivo, destino) 
```

```python
#Questão 6

# adicionando  no dicionario 
dict_diretorio["outro"] = {"caminho": "./arquivo_eleicao/outro"}
for chave, valor in dict_diretorio.items():
    print(valor)

# adicionando pasta outro de acordo com o dicionario 
for chave, valor in dict_diretorio.items():
   os.makedirs(valor["caminho"], exist_ok=True)
   if valor["caminho"] != f'./arquivo_eleicao/outro':
      os.makedirs(f'{valor["caminho"]}/outro', exist_ok=True)
      
#movendo arquivos para pasta outro      

def arquivo_leiame(file):
    return (file.split('_')[-1].replace('.pdf', "") == 'leiame' or
            file.split('_')[0].replace('.pdf', "") == 'leiame')

def arquivo_brasil(file):
    return file.split('_')[-1].replace('.csv', "") == 'BRASIL'

def mover_arquivo(caminho_arquivo, destino):
    shutil.move(caminho_arquivo, destino)
    print(f"Arquivo {os.path.basename(caminho_arquivo)} movido para {destino}")

for root, dirs, files in os.walk(caminho_inicial):
    for file in files:
        caminho_arquivo = os.path.join(root, file)  

    
        if arquivo_leiame(file) or arquivo_brasil(file):
           
            if 'outro' in dirs:
                dir_destino = dict_diretorio.get('outro', {}).get('caminho')
                
                if dir_destino:  
                    if arquivo_leiame(file):
                        destino = os.path.join(dir_destino, file)
                    else:
                        destino = os.path.join(root, 'outro', file)
                    
                    mover_arquivo(caminho_arquivo, destino)
                else:
                    print("Diretório 'outro' não encontrado no dicionário.")
```
